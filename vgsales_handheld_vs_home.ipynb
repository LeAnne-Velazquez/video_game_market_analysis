##### Identify unique platforms

```python
unique_platforms = sorted(vgsales["Platform"].unique())
print(unique_platforms)
```

##### Drop missing years and convert year to integer

```python
vgsales = vgsales.dropna(subset=["Year"])

vgsales["Year"] = vgsales["Year"].astype(int)
```

```python
vgsales.isna().sum()
```

##### Replace missing publishers w/ null

```python
vgsales["Publisher"] = vgsales["Publisher"].replace("N/A", pd.NA)
```

##### Filter for years 1995-2015

```python
vgsales = vgsales[(vgsales["Year"] >= 1995) & (vgsales["Year"] <= 2015)]
```

##### Excluding PS4 & XOne because they were released toward the end of the 1995-2015 timeframe

```python
vgsales = vgsales[~vgsales["Platform"].isin(["PS4", "XOne"])]
```

```python
vgsales
```

##### Filter 3d platforms

```python
home_3d = ["3DO","DC","GC","N64","PS","PS2","PS3","SAT","Wii","WiiU","XB","X360","PC"]
handheld_3d = ["3DS","PSP","PSV","DS"]

vgsales= vgsales[vgsales["Platform"].isin(home_3d + handheld_3d)]
```

##### Create new column indicating device type

```python
vgsales["Device_Type"] = np.where(
    vgsales["Platform"].isin(handheld_3d),
    "Handheld",
    "Home"
)
```

```python
vgsales
```

##### save df

```python
vgsales.to_csv("../data/vgsales_h_hh_3d.csv", index=False)
```
